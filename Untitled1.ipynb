```python
import folium
import pandas as pd
import json

geo_data_path = "/content/district-boundary-hcm-city.geojson"

try:
    with open(geo_data_path, 'r', encoding='utf-8') as f:
        geo_data = json.load(f)

    districts = [f['properties'].get('name', f['properties'].get('District', f'District {i}'))
                 for i, f in enumerate(geo_data['features'])]
    df = pd.DataFrame({
        'Ten khu vuc': districts,
        'Doanh thu': [i * 10 + 100 for i in range(len(districts))]
    })

    m = folium.Map(location=[10.776, 106.700], zoom_start=11)

    folium.Choropleth(
        geo_data=geo_data,
        data=df,
        columns=["Ten khu vuc", "Doanh thu"],
        key_on="feature.properties.name",
        fill_color="YlOrRd",
        fill_opacity=0.7,
        line_opacity=0.2,
        legend_name="Doanh thu (triệu đồng)"
    ).add_to(m)

    display(m)

except Exception as e:
    print(f"Lỗi: {e}")
```

```python

import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

place = " District 1, Ho Chi Minh City , Vietnam"
G = ox.graph_from_place(place, network_type='drive')

print("Đã tải dữ liệu mạng giao thông!")

print("Hiển thị bản đồ...")
ox.plot_graph(G)


# a. Số nút giao
num_nodes = len(G.nodes)
print("Số nút giao:", num_nodes)

# b. Chiều dài đường trung bình
lengths = [data['length'] for u, v, data in G.edges(data=True)]
avg_length = np.mean(lengths)
print("Chiều dài đường trung bình:", avg_length, "m")

nodes = list(G.nodes)

# chọn 2 điểm bất kỳ
orig = nodes[0]
dest = nodes[100]

route = nx.shortest_path(G, orig, dest, weight='length')

print("Đã tìm được đường đi ngắn nhất!")

# Vẽ đường đi
ox.plot_graph_route(G, route)

print("Mô phỏng tắc nghẽn...")

for u, v, data in G.edges(data=True):
    if data['length'] > 100:  # giả lập đường dài bị kẹt
        data['length'] *= 2

# Tính lại đường đi
new_route = nx.shortest_path(G, orig, dest, weight='length')

print("Đường đi sau khi tắc nghẽn:")

# Vẽ lại
ox.plot_graph_route(G, new_route)
```

```python
import osmnx as ox
import networkx as nx
import folium
import random
import math


place = "District 1 , Ho Chi Minh City, Vietnam"
G = ox.graph_from_place(place, network_type='drive')

center = [10.7769, 106.7009]
m = folium.Map(location=center, zoom_start=14)


def random_point():
    return (
        center[0] + random.uniform(-0.01, 0.01),
        center[1] + random.uniform(-0.01, 0.01)
    )

cars = [random_point() for _ in range(5)]
customers = [random_point() for _ in range(5)]


def distance(p1, p2):
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)

assignments = []

for cust in customers:
    best_car = None
    min_dist = float('inf')

    for car in cars:
        d = distance(cust, car)
        if d < min_dist:
            min_dist = d
            best_car = car

    assignments.append((cust, best_car))


# Vẽ xe
for car in cars:
    folium.Marker(location=car, icon=folium.Icon(color='blue'), popup="Car").add_to(m)

# Vẽ khách
for cust in customers:
    folium.Marker(location=cust, icon=folium.Icon(color='red'), popup="Customer").add_to(m)

# Vẽ đường nối
for cust, car in assignments:
    folium.PolyLine([cust, car], color='green').add_to(m)

m
```

```python
import folium
import itertools
from geopy.distance import geodesic

warehouse = (10.776, 106.700)

customers = {
    "Khách 1": (10.780, 106.710),
    "Khách 2": (10.770, 106.690),
    "Khách 3": (10.765, 106.705)
}

random_route = [warehouse] + list(customers.values())

random_dist = sum(
    geodesic(random_route[i], random_route[i+1]).km
    for i in range(len(random_route)-1)
)


best_route = None
min_dist = 1e9

for perm in itertools.permutations(customers.items()):
    route = [warehouse] + [p[1] for p in perm]

    dist = sum(
        geodesic(route[i], route[i+1]).km
        for i in range(len(route)-1)
    )

    if dist < min_dist:
        min_dist = dist
        best_route = route

print("Khoảng cách không tối ưu:", round(random_dist,2), "km")
print("Khoảng cách tối ưu:", round(min_dist,2), "km")

m = folium.Map(location=warehouse, zoom_start=13)

# Kho
folium.Marker(warehouse, popup="Kho", icon=folium.Icon(color="red")).add_to(m)

# Khách
for name, loc in customers.items():
    folium.Marker(loc, popup=name, icon=folium.Icon(color="blue")).add_to(m)

# Tuyến không tối ưu
folium.PolyLine(random_route, color="black", tooltip="Không tối ưu").add_to(m)

# Tuyến tối ưu (màu xanh)
folium.PolyLine(best_route, color="blue", tooltip="Tối ưu").add_to(m)

m
```
