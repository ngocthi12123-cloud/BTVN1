```python
#23.7
import osmnx as ox
import networkx as nx

place = "District 1, Ho Chi Minh City, Vietnam"
G = ox.graph_from_place(place, network_type='drive')

print("--- TÌM ĐƯỜNG ĐI TỐI ƯU (BÀI 23.7) ---")

# 2. Chọn 2 điểm bất kỳ (Điểm đầu và điểm cuối)
nodes = list(G.nodes)
orig = nodes[0]   # Điểm A
dest = nodes[100] # Điểm B

# 3. Tìm đường đi ngắn nhất bằng thuật toán Dijkstra
route = nx.shortest_path(G, orig, dest, weight='length')
print("Đã tìm được đường đi ngắn nhất (Dijkstra)!")
ox.plot_graph_route(G, route)

# 4. MÔ PHỎNG TẮC NGHẼN
# Giả sử các đoạn đường dài > 100m đang bị kẹt xe, ta tăng "trọng số" thời gian đi qua lên gấp đôi
for u, v, data in G.edges(data=True):
    if data['length'] > 100:
        data['length'] *= 2

# 5. Tính lại đường đi sau khi có tắc nghẽn
new_route = nx.shortest_path(G, orig, dest, weight='length')
print("Đường đi mới đã được đề xuất để tránh các đoạn tắc nghẽn:")
ox.plot_graph_route(G, new_route)

# Ý nghĩa quản trị: Đây là lõi của các ứng dụng như Google Maps.
# AI giúp tài xế tránh vùng kẹt xe bằng cách cập nhật dữ liệu đường đi liên tục theo thời gian thực.
```

```python
#23.8
import osmnx as ox
import networkx as nx
import folium
import random
import math


place = "District 1 , Ho Chi Minh City, Vietnam"
G = ox.graph_from_place(place, network_type='drive')

center = [10.7769, 106.7009]
m = folium.Map(location=center, zoom_start=14)

def random_point():
    return (
        center[0] + random.uniform(-0.01, 0.01),
        center[1] + random.uniform(-0.01, 0.01)
    )

cars = [random_point() for _ in range(5)]
customers = [random_point() for _ in range(5)]

def distance(p1, p2):
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)

assignments = []

for cust in customers:
    best_car = None
    min_dist = float('inf')

    for car in cars:
        d = distance(cust, car)
        if d < min_dist:
            min_dist = d
            best_car = car

    assignments.append((cust, best_car))

# Vẽ xe
for car in cars:
    folium.Marker(location=car, icon=folium.Icon(color='blue'), popup="Car").add_to(m)

# Vẽ khách
for cust in customers:
    folium.Marker(location=cust, icon=folium.Icon(color='red'), popup="Customer").add_to(m)

# Vẽ đường nối
for cust, car in assignments:
    folium.PolyLine([cust, car], color='green').add_to(m)

m
```

```python
#23.12
import folium
import itertools
from geopy.distance import geodesic

warehouse = (10.776, 106.700)

customers = {
    "Khách 1": (10.780, 106.710),
    "Khách 2": (10.770, 106.690),
    "Khách 3": (10.765, 106.705)
}

random_route = [warehouse] + list(customers.values())

random_dist = sum(
    geodesic(random_route[i], random_route[i+1]).km
    for i in range(len(random_route)-1)
)


best_route = None
min_dist = 1e9

for perm in itertools.permutations(customers.items()):
    route = [warehouse] + [p[1] for p in perm]

    dist = sum(
        geodesic(route[i], route[i+1]).km
        for i in range(len(route)-1)
    )

    if dist < min_dist:
        min_dist = dist
        best_route = route

print("Khoảng cách không tối ưu:", round(random_dist,2), "km")
print("Khoảng cách tối ưu:", round(min_dist,2), "km")

m = folium.Map(location=warehouse, zoom_start=13)

# Kho
folium.Marker(warehouse, popup="Kho", icon=folium.Icon(color="red")).add_to(m)

# Khách
for name, loc in customers.items():
    folium.Marker(loc, popup=name, icon=folium.Icon(color="blue")).add_to(m)

# Tuyến không tối ưu
folium.PolyLine(random_route, color="black", tooltip="Không tối ưu").add_to(m)

# Tuyến tối ưu (màu xanh)
folium.PolyLine(best_route, color="blue", tooltip="Tối ưu").add_to(m)

m
```

```python
#23.13
import folium

# 1. Khởi tạo bản đồ trung tâm TP.HCM
dashboard = folium.Map(location=[10.776, 106.701], zoom_start=13, control_scale=True)

# 2. TẠO LỚP DỮ LIỆU: ĐIỂM (CÁC CỬA HÀNG)
layer_stores = folium.FeatureGroup(name='Hệ thống Cửa hàng (Điểm)')
stores = [
    {"loc": [10.776, 106.701], "name": "Chi nhánh Quận 1"},
    {"loc": [10.785, 106.665], "name": "Chi nhánh Quận 3"},
    {"loc": [10.758, 106.675], "name": "Chi nhánh Quận 5"}
]
for s in stores:
    folium.Marker(
        location=s['loc'],
        popup=s['name'],
        icon=folium.Icon(color='blue', icon='shopping-cart', prefix='fa')
    ).add_to(layer_stores)

# 3. TẠO LỚP DỮ LIỆU: VÙNG (PHẠM VI PHỤC VỤ 2KM)
layer_zones = folium.FeatureGroup(name='Vùng phủ sóng (Vùng)')
for s in stores:
    folium.Circle(
        location=s['loc'],
        radius=2000, # 2000 mét
        color='orange',
        fill=True,
        fill_opacity=0.2,
        popup=f"Vùng giao hàng 2km của {s['name']}"
    ).add_to(layer_zones)

# 4. TẠO LỚP DỮ LIỆU: TUYẾN ĐƯỜNG (LỘ TRÌNH TIẾP TẾ)
layer_routes = folium.FeatureGroup(name='Tuyến vận chuyển (Đường)')
# Giả sử xe đi từ Q1 -> Q3 -> Q5
route_coords = [s['loc'] for s in stores]
folium.PolyLine(
    locations=route_coords,
    color='red',
    weight=4,
    opacity=0.7,
    tooltip="Tuyến xe tải tiếp hàng hàng ngày"
).add_to(layer_routes)

# 5. KẾT HỢP VÀO DASHBOARD VÀ THÊM TÍNH TƯƠNG TÁC
layer_stores.add_to(dashboard)
layer_zones.add_to(dashboard)
layer_routes.add_to(dashboard)

# Thêm nút bật/tắt các lớp dữ liệu (Đây là tính năng tương tác quan trọng của Dashboard)
folium.LayerControl(collapsed=False).add_to(dashboard)

# Hiển thị Dashboard
dashboard
```

```python
#23.4
import folium
import pandas as pd
import json

geo_data_path = "/content/district-boundary-hcm-city (1).geojson"
excel_data_path = "/content/TTT5.xlsx"

try:
    with open(geo_data_path, 'r', encoding='utf-8') as f:
        geo_data = json.load(f)
    df = pd.read_excel(excel_data_path)

    def fix_name(x):
        x = str(x).strip()
        if "Quan " in x:
            return x.replace("Quan ", "District ")
        return x

    df['Ten_Mapping'] = df['Ten khu vuc'].apply(fix_name)


    m = folium.Map(location=[10.776, 106.700], zoom_start=11)

    folium.Choropleth(
        geo_data=geo_data,
        data=df,
        columns=["Ten_Mapping", "Doanh thu"],
        key_on="feature.properties.name",
        fill_color="YlOrRd",
        fill_opacity=0.7,
        line_opacity=0.5,
        nan_fill_color="brey", # Quận nào không có dữ liệu sẽ hiện màu xám nhạt
        legend_name="Doanh thu (triệu đồng)"
    ).add_to(m)

    revenue_dict = df.set_index('Ten_Mapping')['Doanh thu'].to_dict()

    for feature in geo_data['features']:
        name_in_geo = feature['properties'].get('name', '')

        val = revenue_dict.get(name_in_geo, 0)
        feature['properties']['doanh_thu_hien_thi'] = val

    folium.GeoJson(
        geo_data,
        style_function=lambda x: {'fillColor': 'transparent', 'color': 'black', 'weight': 0.5},
        tooltip=folium.GeoJsonTooltip(
            fields=['name', 'doanh_thu_hien_thi'],
            aliases=['Quận/Huyện:', 'Doanh thu:'],
            localize=True
        )
    ).add_to(m)

    display(m)

except Exception as e:
    print(f"Lỗi: {e}")
```
